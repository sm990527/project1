```python
"""교통사고"""
#1 
import pandas as pd
acci=pd.read_csv('acci.csv',engine='python')
acci.head()
```

```python
#2
group_names = ['적다','보통','많다']
acci['frequency'],mybin = pd.cut(acci['Occurrence'],3,labels=group_names,retbins=True)
acci
```

```python
#3
acci = pd.concat([acci, pd.get_dummies(acci['frequency'])],axis=1,sort=False)
acci
```

```python
#4
acci.isnull().sum()
```

```python
#5
acci['Occurrence']=(acci['Occurrence']-acci['Occurrence'].min())/(acci['Occurrence'].max()-acci['Occurrence'].min())
acci['death']=(acci['death']-acci['death'].min())/(acci['death'].max()-acci['death'].min())
acci['hurt']=(acci['hurt']-acci['hurt'].min())/(acci['hurt'].max()-acci['hurt'].min())
acci['heavy']=(acci['heavy']-acci['heavy'].min())/(acci['heavy'].max()-acci['heavy'].min())
acci['light']=(acci['light']-acci['light'].min())/(acci['light'].max()-acci['light'].min())
acci['report']=(acci['report']-acci['report'].min())/(acci['report'].max()-acci['report'].min())
acci
```

```python
#6-1 결정트리
import numpy as np
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

x=np.array(pd.DataFrame(acci,columns=['Occurrence','death','hurt','heavy','light','report']))
y=np.array(pd.DataFrame(acci,columns=['law']))

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3)

clf=tree.DecisionTreeClassifier()
clf.fit(x_train,y_train)

predictions=clf.predict(x_test)

print("예측:",predictions)
print("결과(y_test):",y_test)
print(accuracy_score(y_test, predictions))
```

```python
#6-2 K최근접 이웃
feature_names = acci[['Occurrence','death','hurt','heavy','light','report']]
feature = np.array(feature_names)
target_names = acci['law']
names = np.array(target_names)
x = feature
y = names

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.3)

from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
clf.fit(x_train, y_train)
predictions = clf.predict(x_test)
print('예측:\n', predictions)
print('정답(y_test):\n', y_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictions))
```

```python
# 6-3 서포트 벡터 머신
feature_names = acci[['Occurrence','death','hurt','heavy','light','report']]
feature = np.array(feature_names)
target_names = acci['law']
names = np.array(target_names)
X = feature
y = names

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.3)

from sklearn import svm
clf = svm.SVC(gamma=0.001, C=100.)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
print('예측:\n', predictions)
print('정답(y_test):\n', y_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictions))
```
